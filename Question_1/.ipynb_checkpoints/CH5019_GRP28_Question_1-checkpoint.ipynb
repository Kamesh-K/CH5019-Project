```python
# Program to perform image classification based on facial recognition
# Importing the necessary libraries for the program
import numpy as np
import os
import glob
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Importing the images from the given data 
# The images are stored in the Dataset_Question1 Directory
pwd = os.getcwd()
data_wd = pwd+"/CH5019_GRP28_Dataset_Question1/"
batch_size = 10   # Represents how many images of a particular person is there -- Referred to as batch
total_people = 15 # Total number of people present to perform the classification 
total_images = batch_size*total_people # Total number of images available
d1 = 64 # Dimensions of the images are given by d1 and d2
d2 = 64
F = np.zeros((d1, d2, total_images)) # Storing all the images in the variable F 
for batch in range(total_people):
    # All the images are loaded using glob and plt.imread and the data is stored in variable F 
    os.chdir(data_wd+str(batch+1))
    names = sorted(glob.glob("*.pgm"))
    names = sorted(names,key=lambda x: int(os.path.splitext(x)[0]))
    d1, d2 = plt.imread(names[0]).shape
    num = len(names)
    for n, i in enumerate(names):
        F[:, :, batch*batch_size+n] = (plt.imread(i,0).astype(np.double)) / 255.
    os.chdir(pwd)
# The variable is reshaped so that an image is represented as a column in the matrix F 
F = F.reshape(d1 * d2, batch_size * total_people)
clip = 10
print("The dimensions of the Faces Matrix F = {}".format(F.shape))
os.chdir(pwd)
```

```python
# Constructing the representative images from the given set of images for each of the individuals
RepresentativeImages = np.zeros((d1*d2,total_people))
# Number of components to be taken to reconstruct the images to calculate the representative image of a person
Components = 1
for people in range(total_people):
    # Storing the face images of a person in the variable F_person 
    F_person = F[:,people*batch_size:(people+1)*batch_size]
    # Performing SVD on the F_person variable to obtain the base faces to perform reconstruction using chosen components
    u,s,vh = np.linalg.svd(F_person)
    sigma_matrix = s[:Components]*np.identity(s[:Components].shape[0])
    u_selected=u[:, :Components]
    vh_selected=vh[:Components, :]
    # All the images of the person are reconstructed using the components obtained from SVD 
    reconstructed_images = np.dot(u_selected, np.dot(sigma_matrix, vh_selected))
    # The representative image is chosen to be the average image of the images reconstructed for the given person using the chosen components
    RepresentativeImages[:,people] = np.mean(reconstructed_images,axis=1)
```

```python
# Representative images of the people
fig, ax = plt.subplots(nrows=3, ncols=5)
fig.suptitle("Representative Images of People",fontsize = 16)
axn = ax.flatten()
PeopleID = 0 
for axes in axn:
    axes.imshow(np.reshape(RepresentativeImages[:,PeopleID],(d1,d2)))
    PeopleID += 1 
    axes.axis('off')
    axes.title.set_text("Person "+str(PeopleID))
plt.show()
```

```python
# Function to test the face recognition for the image file who's filename is passed as an argument
def test_FaceReg(FileName, flag=0): 
    # Reading the image with the particular file name and performing the testing process
    test_image = (plt.imread(str(FileName)+".pgm",0).astype(np.double)) / 255.
    test_image = np.reshape(test_image,(d1*d2))
    err_arr = []
    # Calculating the norm difference between the test image and the representative images of people
    # The identity of the representative image with minimal difference is assigned to the test image 
    for people in range(total_people):
        error = np.linalg.norm(test_image-RepresentativeImages[:,people])
        err_arr.append(error)
    Identity = err_arr.index(min(err_arr)) + 1
    # If the flag passed is 1, we output the test image and the representative image of the person predicted
    if flag == 1:
        fig, ax = plt.subplots(nrows=1, ncols=2)
        fig.suptitle("Test output",fontsize = 16)
        axn = ax.flatten()
        axn[0].imshow(np.reshape(test_image,(d1,d2)))
        axn[0].title.set_text("Input Image")
        axn[1].imshow(np.reshape(RepresentativeImages[:,Identity-1],(d1,d2)))
        axn[0].axis('off')
        axn[1].axis('off')
        axn[1].title.set_text("Person "+str(Identity))
        plt.show()
    return Identity
```

```python
# Array to store all the predictions of the dataset 
Predictions = []
for people in range(total_people):
    os.chdir(data_wd+str(people+1))
    for batch in range(batch_size):
        Predictions.append(test_FaceReg(str(batch+1)))
os.chdir(data_wd)
Predictions = np.asarray(Predictions)
# Storing the actual results which are supposed to compare 
Result = np.zeros((total_people, batch_size))
for people in range(total_people):
    Result[people,:] = people + 1
Result = np.reshape(Result,(total_images))
# Assigning the score based on whether the predicted class is equal to the actual class
Score = np.equal(Predictions,Result)
Score = np.sum(Score)
# The score of the model is printed
print("Number of images correctly mapped out of 150 images = {} and hence the score = {}".format(Score,Score*100/150))
```

```python
# Example to show how the output is represented for a given test image, the directory is the identity of the person
# The argument passed to the function is the filename and whether to print the image output or not 
os.chdir(data_wd+'/2')
test_FaceReg(2,1)
```
